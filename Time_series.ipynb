# Data

```python
import numpy as np
import matplotlib.pyplot as plt

n = 60
t = np.arange(n)
beta0 = 100
beta1 = 2
A = 20
B = 12
sigma = 5

trend = beta0 + beta1 * t
seasonality = A * np.sin(2 * np.pi * t / B)
noise = np.random.normal(0, sigma, n)

time_series = trend + seasonality + noise

plt.figure(figsize=(12, 6))
plt.plot(t, time_series)
plt.xlabel("Time (months)")
plt.ylabel("Value")
plt.title("Generated Time Series Data")
plt.grid(True)
plt.plot(t, trend, label="Trend")
plt.plot(t, seasonality, label="Seasonality")
plt.plot(t, noise, label="Noise")
plt.legend()
plt.show()
```

# Model regression lineaire

```python
# Custom implementation of Simple Linear Regression
def simple_linear_regression(x, y):
    n = len(x)
    sum_x = sum(x)
    sum_y = sum(y)
    sum_xy = sum(x_i * y_i for x_i, y_i in zip(x, y))
    sum_x_squared = sum(x_i**2 for x_i in x)

    # Calculate coefficients
    beta1_custom = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x**2)
    beta0_custom = (sum_y - beta1_custom * sum_x) / n

    return beta0_custom, beta1_custom

# Get the coefficients from the custom model
beta0_custom, beta1_custom = simple_linear_regression(t, time_series)

# Generate predictions using the custom model
naive_predictions_custom = beta0_custom + beta1_custom * t

# Print the custom model coefficients
print(f"Custom Naive Model - Intercept (beta0): {beta0_custom:.2f}")
print(f"Custom Naive Model - Slope (beta1): {beta1_custom:.2f}")
```

```python
# Plot the original time series and the custom naive model's predictions
plt.figure(figsize=(12, 6))
plt.plot(t, time_series, label="Original Time Series")
plt.plot(t, naive_predictions_custom, label="Custom Naive Model Predictions", color='green')
plt.xlabel("Time (months)")
plt.ylabel("Value")
plt.title("Custom Naive Model Regression on Time Series Data")
plt.grid(True)
plt.legend()
plt.show()
```

# Metrics

```python
# Calculate the errors for each time point
errors_custom = time_series - naive_predictions_custom

# Calculate the squared errors for each time point
squared_errors_custom = errors_custom**2

# Print or display the errors and squared errors for each time point
print("Errors and Squared Errors for each time point (Custom Model):")
for i in range(n):
    print(f"Time {t[i]}: Error = {errors_custom[i]:.2f}, Squared Error = {squared_errors_custom[i]:.2f}")

# Note: RMSE and MSE are calculated from the average of the squared errors.
# MAE is calculated from the average of the absolute errors.
# The values shown here are the contributions to those averages at each time point.
```
